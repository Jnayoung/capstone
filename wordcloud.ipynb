```python
a = open("test.txt","r", encoding = 'utf-8')
```

```python
a = a.readlines()
A = []
for x in a:
    A.append(x.strip())
```

```python
import re
```

```python
AA=[]
for x in A:
     AA.append(re.sub('[^가-힣 ]', '', x))
```

```python
## 3_1 : 공집합 제거

for x in AA:
    if x == "":
        AA.remove(x)
    if x == " ":
        AA.remove(x)
    if x == "  ":
        AA.remove(x)
## 3_2 : 양 옆 공백 제거 + 새로운 집합 탄생

AAA = []
for x in AA:
    AAA.append(x.strip())
```

```python
from konlpy.tag import Komoran
komoran = Komoran()
```

```python
AAA2 = []

for x in AAA:
    AAA2.append(re.sub('[상암+점+]','',x))
```

```python
stopwords = ['수','것','점','집','때','거','원','정도','식','번','중','분','와','안', '명', '일부',
             '결', '자', '월', '판','말','다음','살','등','입','감','한','위','시','게','건','이랑','제가',
             '듯','글','이','회','후','가지','때문','포','편','육','속','토','곳','일',
            '인','심','앞','야','전','층','장','간','알','보니','개','날','그런지','삼','지','바','데']
```

```python
# 명사 추출
noun_list = []

for x in AAA2:
    if x =='':
        continue
    else:
        #n = komoran.nouns(x)
        for y in komoran.nouns(x):
            if not y in stopwords:
                noun_list.append(y)

noun_list
```

```python
import nltk
from nltk import FreqDist
import numpy as np
vocab = FreqDist(np.hstack(noun_list))
# 5_1 100개만 뽑기
vocab2 = vocab.most_common(100)
vocab2
```

```python
import numpy as np
from PIL import Image

mask = Image.new("RGBA",(2555,2275), (255,255,255))
image = Image.open('cloud.png').convert("RGBA")
x,y = image.size
mask.paste(image,(0,0,x,y),image)
mask = np.array(mask)
# 7 shape 이미지틀에서 빈도수별로 크기가 다른 단어들 출력하기
from wordcloud import WordCloud

font='12롯데마트행복Bold.ttf'
wc = WordCloud(font_path=font, background_color='white',
               width = 700, height = 700, max_font_size=400,
               mask=mask,colormap='Dark2_r').generate_from_frequencies(dict(vocab2))

# 사이즈 설정 및 화면에 출력
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.imshow(wc)
plt.axis('off')
plt.savefig('result.png', bbox_inches='tight', pad_inches=0.5)
plt.show()
```
